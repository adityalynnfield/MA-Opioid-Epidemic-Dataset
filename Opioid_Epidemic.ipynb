```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Combined data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.replace(',','', regex=True)
data = data.replace('%','', regex=True)
```

```python
data.head()
```

```python
convert_dict = {'Population': int,
                'DOR Income': int,
                'DOR Income Per Capita': int,
                'EQV': int,
                'EQV Per Capita': int,
                'Democrat': float,
                'Republican': float
                }
```

```python
data = data.astype(convert_dict)
```

```python
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data["Opiod Deaths"], bins = 30, kde=True, color='orange', stat="count", edgecolor='black')


plt.title("Distribution of Opiod Deaths")
plt.xlabel("Opiod Deaths")
plt.ylabel("Count")


plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data["DOR Income Per Capita"], bins = 30, kde=True, color='orange', stat="count", edgecolor='black')

plt.xlim(0, 400000)


plt.title("Distribution of Income Per Capita")
plt.xlabel("Income Per Capita")
plt.ylabel("Count")


plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data["Unemployment"], bins = 30, kde=True, color='orange', stat="count", edgecolor='black')


plt.title("Distribution of Unemployment")
plt.xlabel("Unemployment")
plt.ylabel("Count")


plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data["Democrat"], bins = 30, kde=True, color='orange', stat="count", edgecolor='black')


plt.title("Distribution of Democratic Votes")
plt.xlabel("Democrat")
plt.ylabel("Count")


plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data["Republican"], bins = 30, kde=True, color='orange', stat="count", edgecolor='black')


plt.title("Distribution of Republican Votes")
plt.xlabel("Republican")
plt.ylabel("Count")


plt.show()
```

```python


x = data["DOR Income Per Capita"]
y = data["Opiod Deaths"]


plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='orange', alpha=0.6,  marker='x')


plt.title("Opioid Deaths vs Income Per Capita")
plt.xlabel("DOR Income Per Capita")
plt.ylabel("Opioid Deaths")

# Set axis limits (adjust based on your data)
plt.xlim(0, 400000)
plt.ylim(0, 400)


plt.show()
```

```python
x = data["Unemployment"]
y = data["Opiod Deaths"]


plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='orange', alpha=0.6,  marker='x')

plt.title("Opioid Deaths vs Unemployment")
plt.xlabel("Unemployment")
plt.ylabel("Opioid Deaths")


plt.show()
```

```python
x = data["Democrat"]
y = data["Opiod Deaths"]


plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='orange', alpha=0.6,  marker='x')

plt.title("Opioid Deaths vs Democrat Vote Percentage")
plt.xlabel("Democrat")
plt.ylabel("Opioid Deaths")



plt.show()
```

```python
x = data["Population"]
y = data["Opiod Deaths"]


plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='orange', alpha=0.6,  marker='x')

plt.title("Opioid Deaths vs Population of Town")
plt.xlabel("Population")
plt.ylabel("Opioid Deaths")



plt.show()
```

```python
correlation_matrix = data[["Opiod Deaths", "DOR Income Per Capita", "Unemployment", "Democrat", "Republican"]].corr()


plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
